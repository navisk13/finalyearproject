```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
#read the data into a dataframe
train_df=pd.read_csv('freqdata.csv', header=None)
print(train_df)
```

```
#store the target column(rate) in a separate array
rate=np.array(train_df[[train_df.columns[225]]])
print(rate)
```

```
#assign classes according to the respiratory rate
actual_class=[]
for x in rate:
  if x==0:
    actual_class.append(0)
  elif x>0 and x<40:
    actual_class.append(1)
  elif x>=40 and x<=60:
    actual_class.append(2)
  elif x>60:
    actual_class.append(3)
```

```
#print the class labels we assigned
actual_class=np.vstack(actual_class)
print(actual_class)
```

```
#split the data into training and testing, with 20% for testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_df, actual_class, test_size=0.20)  
print(X_test)
```

```
#store the rates for training and testing data in separate arrays
rate_train=np.array(X_train[[X_train.columns[225]]])
rate_test=np.array(X_test[[X_test.columns[225]]])
```

```
#remove the rate target column from training data
X_train=X_train.drop(X_train.columns[225],axis=1);
print(X_train)
```

```
#remove the rate target column from testing data
X_test=X_test.drop(X_test.columns[225],axis=1);
print(X_test)
```

```
from keras.utils import to_categorical
#one-hot encode target column
train_y_2 = to_categorical(y_train, num_classes=4, dtype='float32')

#vcheck that target column has been converted
train_y_2[0:5]
```

```
from keras.models import Sequential
from keras.layers import Dense
#create model
model = Sequential()

#get number of columns in training data
n_cols = X_train.shape[1]

#add layers to model
model.add(Dense(n_cols, activation='relu', input_shape=(n_cols,)))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(4, activation='softmax'))

#train the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, train_y_2, epochs=500)
```

```
#predict and print the class labels for testing data.
y_class=model.predict(X_test)
classes=np.argmax(y_class,axis=1)
print(classes)
```

```
#add the classes along with their corresponding respiratory rates in a separate dataframe
classes=np.vstack(classes)
rate_df=pd.DataFrame(rate_test,columns=["X"])
classes_df=pd.DataFrame(classes,columns=["Y"])
data = pd.concat([rate_df,classes_df],axis=1)
print(data)
```

```
#split the dataframe into 4 according to classes. Rates for each class will be in a distinct dataframe
df1 = data[data['Y'] == 0]
df2 = data[data['Y'] == 1]
df3 = data[data['Y'] == 2]
df4 = data[data['Y'] == 3]
```

```
#scatter plot the rates and their classes. Rates are on the x axis, and classes on the y axis.
ax=df1.plot.scatter(x='X',y='Y',c='r', label="No Information")
df2.plot.scatter(ax=ax,x='X',y='Y',c='b', label="Bradypnea")
df3.plot.scatter(ax=ax,x='X',y='Y',c='g', label="Healthy")
plt.legend()
plt.show()
```

```
#generate a confusion matrix for the testing data.
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,classes))
```
